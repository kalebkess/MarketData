```python
import pandas as pd
import requests

def get_instruments():
  instruments = []
  target_url = 'https://api.robinhood.com/instruments/'
  i = 1
  while target_url != None:
    print(f'[{i}] fetching {target_url}')
    i += 1
    r = requests.get(target_url).json()
    #instruments.extend([instrument for instrument in r['results']])
    instruments.extend([*r['results']])
    target_url = r['next']
  instruments_df = pd.DataFrame(instruments).set_index('id')
  instruments_df.to_csv('../examples/RH_instruments.csv')
  return instruments_df
```

```python
data = get_instruments()
```
